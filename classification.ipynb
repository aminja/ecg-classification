```python
import pandas as pd

data = pd.read_csv('data/mitbih_data.csv', header=None)
data.head()
```

```python
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

kf = KFold(n_splits=5)
clf = RandomForestClassifier(n_estimators=100, max_depth=200, random_state=0)

idx = 0
confusion_matrices = []
for train_index, test_index in kf.split(data):
    train_X, train_y = data.iloc[train_index, :-1], data.iloc[train_index, -1]
    test_X, test_y = data.iloc[test_index, :-1], data.iloc[test_index, -1]
    clf.fit(train_X, train_y)
    y_pred = clf.predict(test_X)
    confusion_matrices.append(confusion_matrix(test_y, y_pred, labels=[0, 1, 2, 3, 4]))
    idx += 1
    print('Accuracy of model in section {}: {}'.format(idx, accuracy_score(test_y, y_pred)))
```

```python
def get_measures(confusion_matrix):
    tp = confusion_matrix.trace() - confusion_matrix[0,0]
    fn = confusion_matrix[1:, 0].sum()
    fp = confusion_matrix[0, 1:].sum()
    return tp, fn, fp

for c in confusion_matrices:
    print(c)
    tp, fn, fp = get_measures(c)
    recall = tp / (tp + fn)
    precision = tp / (tp + fp)
    print('Recall: {}'.format(recall))
    print('Precision: {}'.format(precision))
    print('===============================')
    
    
```
