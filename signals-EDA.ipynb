```python
import pandas as pd

data = pd.read_csv('data/mitbih_data.csv', header=None)
data.head()
```

```python
data.info()
data.iloc[:,-1].unique()
```

```python
data.iloc[:, -1].value_counts().plot.bar(x='ECG Class', y='Frequency')
```

```python
# Plot abnormal ECG signals
import matplotlib.pyplot as plt
# %matplotlib inline

fig, axs = plt.subplots(5, figsize=(7,15))
# fig.suptitle('Vertically stacked subplots')

for c in range(0, 5):
    signal = data[data.iloc[:, -1] == c].head(1)
    axs[c].plot(signal.iloc[:, :-1].T, label='Class-{}'.format(c))
    axs[c].legend(loc='upper right')
```

```python
# read top 3 normal signals
ax = data.iloc[0:3, :-1].T.plot()
ax.legend(['ECG-0', 'ECG-1', 'ECG-2'])
```
